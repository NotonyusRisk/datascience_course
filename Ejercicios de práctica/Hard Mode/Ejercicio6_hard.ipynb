```python
import pandas as pd
import numpy as np

# Datos ficticios de ventas
data_avanzado = {
    'Fecha': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', 
              '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10'],
    'Cliente': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Edad': [23, 45, 35, 28, 40, 50, 33, 24, 36, 29],
    'Género': ['F', 'M', 'M', 'F', 'M', 'F', 'M', 'F', 'F', 'M'],
    'Canal': ['Online', 'Offline', 'Online', 'Offline', 'Online', 'Offline', 'Online', 'Offline', 'Online', 'Offline'],
    'Producto': ['Camiseta', 'Zapatos', 'Pantalón', 'Camiseta', 'Zapatos', 
                 'Camiseta', 'Pantalón', 'Camiseta', 'Zapatos', 'Pantalón'],
    'Categoría': ['Ropa', 'Calzado', 'Ropa', 'Ropa', 'Calzado', 
                   'Ropa', 'Pantalón', 'Ropa', 'Zapatos', 'Ropa'],
    'Cantidad': [1, 2, 1, 3, 1, 2, 1, 2, 3, 1],
    'Precio': [25, 50, 75, 25, 50, 75, 25, 50, 75, 25],
    'Ventas Totales': [25, 100, 75, 75, 50, 150, 100, 50, 150, 75],
    'Región': ['Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur']
}

# Crear el DataFrame
df_avanzado = pd.DataFrame(data_avanzado)

df_avanzado
```

```python
df_avanzado = df_avanzado.dropna()
df_avanzado = df_avanzado.drop_duplicates()
df_avanzado['Fecha'] = pd.to_datetime(df_avanzado['Fecha'])
df_avanzado['Categoría'] = df_avanzado['Categoría'].replace('Zapatos', 'Calzado')
df_avanzado
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_avanzado,
    x='Fecha',
    y='Ventas Totales',
    hue='Producto',
    palette='coolwarm'
)

plt.title('Ventas a lo largo del tiempo por producto')
plt.xlabel('Fechas')
plt.ylabel('Ventas Totales')
plt.legend(title='Producto')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_avanzado,
    x='Fecha',
    y='Ventas Totales',
    hue='Canal',
    palette='coolwarm'
)

plt.title('Ventas a lo largo del tiempo por producto')
plt.xlabel('Fechas')
plt.ylabel('Ventas Totales')
plt.legend(title='Producto')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_avanzado,
    x='Fecha',
    y='Ventas Totales',
    hue='Región',
    palette='coolwarm'
)

plt.title('Ventas a lo largo del tiempo por producto')
plt.xlabel('Fechas')
plt.ylabel('Ventas Totales')
plt.legend(title='Producto')
plt.tight_layout()

plt.show()
```

```python
# Convertir la columna 'Canal' en variable numérica
df_avanzado['Canal_num'] = df_avanzado['Canal'].map({'Online': 1, 'Offline': 0})

# Seleccionar las columnas numéricas para el análisis de correlación
correlacion_df = df_avanzado[['Ventas Totales', 'Edad', 'Cantidad', 'Canal_num']]

# Calcular la matriz de correlación
correlacion = correlacion_df.corr()
correlacion
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df_avanzado,
    x='Fecha',
    y='Ventas Totales',
    palette='coolwarm'
)

plt.title('Ventas en funcion del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas totales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
df_ventas_genero = df_avanzado.groupby('Género')['Ventas Totales'].agg(['sum']).reset_index()
plt.figure(figsize=(10, 6))

sns.barplot(
    data=df_ventas_genero,
    x='Género',
    y='sum'
)

plt.title('Ventas con respecto al género')
plt.xlabel('Géneros')
plt.ylabel('Ventas Totales')
plt.tight_layout()

plt.show()
```

```python
ventas_canal = df_avanzado.groupby('Canal')['Ventas Totales'].agg('sum').reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(
    data=ventas_canal,
    x='Canal',
    y='Ventas Totales'
)

plt.title('Ventas con respecto al canal de ventas')
plt.xlabel('Canal')
plt.ylabel('Ventas Totales')
plt.tight_layout()

plt.show()
```
