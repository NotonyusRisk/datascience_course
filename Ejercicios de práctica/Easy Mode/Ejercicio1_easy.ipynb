```python
import pandas as pd
```

```python
# Crear un DataFrame de ejemplo con datos ficticios de ventas
data = {
    'Fecha': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', 
              '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10'],
    'Producto': ['Camiseta', 'Pantalón', 'Zapatos', 'Camiseta', 'Pantalón', 
                 'Zapatos', 'Camiseta', 'Pantalón', 'Zapatos', 'Camiseta'],
    'Categoría': ['Ropa', 'Ropa', 'Calzado', 'Ropa', 'Ropa', 
                   'Calzado', 'Ropa', 'Ropa', 'Calzado', 'Ropa'],
    'Precio': [25, 50, 75, 25, 50, 75, 25, 50, 75, 25],
    'Cantidad': [1, 2, 1, 3, 1, 2, 4, 1, 2, 3],
    'Ventas Totales': [25, 100, 75, 75, 50, 150, 100, 50, 150, 75]
}

# Crear el DataFrame
df_ventas = pd.DataFrame(data)
df_ventas
```

```python
df_ventas = df_ventas.sort_values(by='Categoría', ascending=True)
df_ventas
```

```python
import matplotlib.pyplot as plt

# Agrupar por categoría y sumas las ventas totales de cada categoría
ventas_per_categoria = df_ventas.groupby('Categoría')['Ventas Totales'].sum()
ventas_per_categoria
```

```python
# Crear un gráfico de barras
plt.figure(figsize=(8, 5))
ventas_per_categoria.plot(kind='bar', color='skyblue')

# Personalizar el gráfico
plt.title('Ventas Totales por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas Totales')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
